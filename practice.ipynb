```python
#5) Write a Python program to find the greatest among three numbers entered by the user
#   without using any built-in functions.

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

greatest = a

if b > greatest:
    greatest = b
if c > greatest:
    greatest = c

print("Greatest number is:", greatest)
```

```python
#6)Write a program to print all numbers between 1 and 50 that are divisible by both 3 and 5
# Also print the count of such numbers.

count = 0

for i in range(1, 51):
    if (i%3==0) and (i%5==0):
        print(i)
        count += 1

print("Count:", count)
```

```python
#8) Write a Python program to take a string as input and count the number of:
#• Vowels
#• Consonants
text = input("Enter a string: ")
vowels = 0
consonants = 0

for i in text:
        if i in 'aeiou':
            vowels += 1
        else:
            consonants += 1

print("No of Vowels:", vowels)
print("No of Consonants:", consonants)
```

```python
#9) Power of Number
#Find a^b using loop (without using **).
a = int(input("Enter number(a): "))
b = int(input("Enter power(b): "))
result = 1
for i in range(b):
    result = result * a

print("Result:", result)
```
