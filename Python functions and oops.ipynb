```python
#18. Write a function to check if a number is a palindrome without converting to string.
def is_palindrome(n):
    if n < 0:
        return False

    digits = []
    while n > 0:
        digits.append(n % 10)
        n //= 10

    arr = np.array(digits)
    return np.array_equal(arr, arr[::-1])
print(is_palindrome(121))
print(is_palindrome(123))
```

Given a string, compress it like:

1.
Input: "aaabbc"
Output: "a3b2c1"


2.
Input: "abcd"
Output: "a1b1c1d1"


3.
Input: "aabbaa"
Output: "a2b2a2"


4.
Input: "aaaaa"
Output: "a5"


5.
Input: "" (empty string)
Output: ""


6.
Input: "aAaA"
Output: "a1A1a1A1"


7.
Input: "aaAAaa"
Output: "a2A2a2"

```python
def compress_string(s):
    res, count = "", 1
    for i in range(1, len(s)+1):
        if i < len(s) and s[i] == s[i-1]:
            count += 1
        else:
            if i > 0:
                res += s[i-1] + str(count)
            count = 1
    return res
```
